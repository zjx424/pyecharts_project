```python
from pyecharts import Bar
import random
X_axis = ["身高","体重","颜值","智商","情商","价值观","人生观"]
bar = Bar(title="条形图示例")
person = ["A", "B", "C", "D", "E"]
for person_name in person:
    bar.add(person_name, X_axis, [random.randint(10, 100) for i in range(7)],mark_line=["average"],mark_point=["max","min"])
bar
```
